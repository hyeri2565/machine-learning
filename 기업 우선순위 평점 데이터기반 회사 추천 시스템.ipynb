```python
import pandas as pd
```

```python
df=pd.read_csv('company_rating_list_new.csv',index_col='index')
```

```python
print(len(df))
df.head()
```

```python
df[df['average']==0]
```

```python
#결측 데이터를 제거한 DataFrame
df=df.drop(df[df['average']==0].index,axis=0)
df
```

```python
df_five=df[['salary','wlb','culture','possibility','management','companyId']]
df_five_index=df_five.set_index('companyId')
df_five_index.head()
len(df_five_index.index.tolist())
```

```python
cus_rate=pd.DataFrame({'rating':[5,3,4,1,2]})
cus_rate
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim_info=[]
for i in range(len(df)):
    sim=cosine_similarity([cus_rate['rating'].tolist()],[df_five.values[i].tolist()])
    sim=sim.tolist()[0]
    sim_info.append(sim[0])
print(sim_info)
```

```python
sim_mt=pd.DataFrame({'similarity':sim_info,'companyId':df_five_index.index.tolist()})
sim_mt=sim_mt.set_index('companyId')
sim_mt_sorted=sim_mt.sort_values(by=['similarity'],ascending=False)
sim_mt_sorted.iloc[:10]
```
